### This is a virtual Pet that interacts with the user on a time-based dropping-energy reality


```python
from random import randrange
from IPython.display import clear_output
import time
```

```python
class Pet():

    excitement_warning = 3
    excitement_max = 10
    food_warning = 2
    food_max = 10
    vocab = ['I want to learn words!', '"sigh*"']
    
    def __init__(self, name):
        self.name = name
        self.excitement = randrange(self.excitement_max)
        self.food = randrange(self.food_max)
        self.vocab = self.vocab[:]
    
    def clock_tick(self):
        self.excitement -= 1
        self.food -= 1
    
    def mood(self):
        if self.excitement > self.excitement_warning and self.food > self.food_warning:
            return "happy"
        elif self.excitement < self.excitement_warning and self.food > self.food_warning:
            return "bored"
        elif self.excitement > self.excitement_warning and self.food < self.food_warning:
            return "hungry"
        else:
            return "sad"
    
    def __str__(self):
        return "I am " + self.name + "." "\nI feel " + self.mood() + "."#" and I  need"
    
    def teach(self, word):
        self.vocab.append(word)
        self.clock_tick()
    
    def talk(self):
        print(f"I am {self.name}. I feel {self.mood()}. \nMy excitement is {self.excitement} and my level of food is {self.food}")
        self.clock_tick()
    
    def feed(self):
        print("***crunch*** \nmmmmm. Thank u!")
        meal = randrange(0, self.food_max)
        self.food += meal
    
    def play(self):
        print("Yeeeeei! <3")
        fun = randrange(0, self.excitement_max)
        self.excitement += fun
        
    def level_parameter(self):
        if self.excitement < 0:
            self.excitement = 0
        elif self.excitement > self.excitement_max:
            self.excitement = self.excitement_max
        
        if self.food < 0:
            self.food = 0
        elif self.food > self.food_max:
            self.food = self.food_max        
    
    def random_chitchat(self):
        print(self.vocab[randrange(len(self.vocab))])
    
    def clean(self):
        pass
```

```python
def main():
    pet_name = input("Name your pet: ")
    my_pet = Pet(pet_name)
    
    start_time = time.time() #to measure the elapsed time and update the minutes counter
    minutes = 0
    talk_to_time = 0 #this is a counter that updates with the minutes counter
    restock_to_time = 0 #this is a counter that updates with the minutes counter
    
    reduce_time = 9 #59 #every n seconds the energy drops (excitement and food) 
    talking_time = 3 # every n minutes the pet says something random form the list of words
    restocking_time = 5 #every n minutes we get +1 in food and play
    
    stock_food = 10 #every feed action get -1 on the stock
    stock_play = 10 #every play action gets -1 on the stock
    
    input("Hello I am " + my_pet.name +" and I am new here!" + "\nPress anything to start")
    choice = None
    clear_output()
    
    while choice != 0:
        
        my_pet.level_parameter()
        
        elapsed_time = round(time.time() - start_time)
        
        if elapsed_time >= reduce_time:
            my_pet.level_parameter()
            minutes = minutes + round(elapsed_time/reduce_time)
            start_time = time.time() + (elapsed_time % reduce_time)
            my_pet.clock_tick()
            print("\n HEY! just to let you know...")
            print(f"I AM {my_pet.mood()}")
        
        if (minutes-talk_to_time) >= talking_time:
            talk_to_time = minutes
            clear_output()
            print("\n***!!!***!!!***!!!***")
            my_pet.random_chitchat()
            print("***!!!***!!!***!!!***\n")
            time. sleep(3)
            clear_output()
        
        if (minutes-restock_to_time) >= restocking_time:
            restock_to_time = minutes
            stock_food += 1
            stock_play += 1
            
        if stock_food < 0:
            stock_food = 0
        if stock_play < 0:
            stock_play = 0
            
        choice = input('''
        Pet Time!
        1 - Feed me
        2 - Talk to me
        3 - Teach me
        4 - Play with me
        0 - Quit(er)
        ''')
        
        clear_output()
        
        if choice == "0":
            print("Good Bye!!!")
            break
        elif choice == "1":
            if stock_food > 0:
                my_pet.feed()
                stock_food -= 1
                continue
            else:
                print("I think we don't have food in the house :/")
                print(f"I AM {my_pet.mood()}")
                time. sleep(2)
                continue
        elif choice == "2":
            my_pet.level_parameter()
            my_pet.talk()
            continue
        elif choice == "3":
            word = input("teach me a new word! ")
            my_pet.teach(word)
            continue
        elif choice == "4":
            if stock_play > 0:
                my_pet.play()
                stock_play -= 1
                continue
            else:
                print("I don't have energy to play :/")
                print(f"I AM {my_pet.mood()}")
                time. sleep(2)
                continue
        elif choice == "I":
            print(f"food: {stock_food} | play: {stock_play}")
            time. sleep(2)
            clear_output()
            continue
        else:
            print("Non valid option. Choose again!")
            continue
            
        clear_output()
```

```python
main()
```
